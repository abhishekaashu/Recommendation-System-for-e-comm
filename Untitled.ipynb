# Content Based Recommendation system

```python
cleaned_data_df.head(1)['Description']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3, max_features=None,
                     strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}',
                     ngram_range=(1,3),
                     stop_words = 'english')

#filling Nans With empty strings
cleaned_data_df['Description'] = cleaned_data_df['Description'].fillna('')
```

```python
#Fitting the Tfidf on the 'Description' text

tfv_matrix = tfv.fit_transform(cleaned_data_df['Description'])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

#compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
sig[0]
```

```python
#Reverse Mapping of indices and  StockCode

indices = pd.Series(cleaned_data_df.index, index=cleaned_data_df['StockCode']).drop_duplicates()
```

```python
def give_rec(Stock_Code, sig=sig):
    #get the index corresponding to original_stockcode
    idx = indices[Stock_Code]
    
    #get the pairwise similarity scores
    sig_scores = list(enumerate(sig[idx]))
    
    #sort the product
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse = True)
    
    #scores of the 10 most similar products
    sig_scores = sig_scores[1:11]
    
    #product indices
    product_indices = [i[0] for i in sig_scores]
    
    #Top 10 most similar product
    return cleaned_data_df['StockCode'].iloc[product_indices]
    
    
    
```
